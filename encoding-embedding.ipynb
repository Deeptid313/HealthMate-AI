```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
tf_idf = TfidfVectorizer()
vector = tf_idf.fit_transform(documents).toarray()
tf_idf
print (vector)
```

```python
for vector in vector:
    print (vector)
```

```python
print ('vocabulary:',tf_idf.get_feature_names_out())
```
