```python
from dotenv import load_dotenv
import os
import torch
from diffusers import StableDiffusionPipeline

load_dotenv(override=True)
token = os.getenv("HUGGINGFACEHUB_API_TOKEN")

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float32,
    use_auth_token=token,
)

pipe = pipe.to("cpu")

image = pipe("A photorealistic portrait of a robot astronaut on Mars").images[0]
image.save("sd1_5.png")
print("Saved sd1_5.png")
```

```python
import base64
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage

##convert image to base64
def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
    
base64_image = image_to_base64("sd1_5.png")
```

```python
##initialize model
model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite"
)
```

```python
##text generation with image input
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "Describe this image in detail."
        },
        {
            "type": "image_url",
            "image_url": {
                "url": f"data:image/png;base64,{base64_image}"
            }
        }
    ]
)
ai_response = model.invoke([message])
print(ai_response.content)
```

```python
from google import genai
from google.genai import types
import wave
```

```python
client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash-preview-tts", #freely accessible model for text to speech conversion
    contents="Hello, Deepti is an AI engineer with expertise in LLMs and GenAI. She is passionate about building AI applications that can solve real-world problems and improve people's lives.'",
    config=types.GenerateContentConfig(
        response_modalities=["AUDIO"],
        speech_config=types.SpeechConfig(
            voice_config=types.VoiceConfig(
                prebuilt_voice_config=types.PrebuiltVoiceConfig(
                    voice_name="Kore"
                )
            )
        )
    )
)
```

```python
audio_data = response.candidates[0].content.parts[0].inline_data.data

with wave.open("output.wav", "wb") as wf:
    wf.setnchannels(1)
    wf.setsampwidth(2)
    wf.setframerate(24000)
    wf.writeframes(audio_data)

print("Audio saved as output.wav")
```

```python
import base64
from pathlib import Path

from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage

def encode_image(image_path: str) -> str:
    image_bytes = Path(image_path).read_bytes()
    return base64.b64encode(image_bytes).decode("utf-8")

image_b64 = encode_image("/Users/rajugaru/Desktop/FullStack-AI/TestDummy/AIbootcampClass/dogs.png")

llm = ChatOllama(
    model="bakllava",
    temperature=0
)

message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "Describe this image in detail."
        },
        {
            "type": "image_url",
            "image_url": f"data:image/jpeg;base64,{image_b64}"
        }
    ]
)

response = llm.invoke([message])

print(response.content)
```
